```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(20)
sinx=np.sin(x)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(x)
ax.plot(sinx)
```

```python
x=np.arange(1,20)
sinx=np.sin(x)
cosx=np.cos(x)
tanx=np.tan(x)
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.plot(x)
ax2=fig.add_subplot(2,2,2)
ax2.plot(sinx)
ax3=fig.add_subplot(2,2,3)
ax3.plot(cosx)
ax4=fig.add_subplot(2,2,4)
ax4.plot(tanx)
```

```python
x = np.arange(1, 20)
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x)
ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
ax2.plot(sinx)
ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
ax3.plot(cosx)
ax4 = fig.add_subplot(2, 2, 4, sharex=ax2, sharey=ax3)
ax4.plot(tanx)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
axes[0, 0].plot(x)
axes[0, 1].plot(sinx)
axes[1, 0].plot(cosx)
axes[1, 1].plot(tanx)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x.cumsum())
ticks = ax.set_xticks([0, 100, 200, 300, 400])
labels = ax.set_xticklabels(['zero', 'quater', 'pre-mid', 'post-mid','final'])
```
